# 4. Selección de variables para el modelo

```python
# Las que influyan en la calificación final.
# Numericas y categoricas ordinales sin influyen en la calificación final o salida
# No se tomara las multiclase, porque parece no influir.

# seleccionar automaticamente las variables. Se pudo haber hecho manual.
list_cat_corrs = 
df_ordinal = dataset_mat[list_cat_corrs].corr()['G3'].reset_index().rename(columns={'index':'variable'})
df_ordinal = df_ordinal[~df_ordinal['variable'].isin(['G3','G2','G1'])] # El rabo de chancho niega, como un not
df_ordinal = df_ordinal[abs(df_ordinal['G3'])>0.21]  # Variables que tenga correlación significativamente alta > 0.21
df_ordinal
list_ordinal_chosen = list(df_ordinal['variable'].values)
list_ordinal_chosen
```

```python
# seleccionar las variables binarias
df_binary = pd.DataFrame()

# Se itera pr las columnas, obteniendo el contraste de hipotesis con G3 y solo nos quedaremos con aquellas que rechacen la 
# hipotesis nula. Que este debajo del umbral Threshold

threshold = 0.05

for column in list_binary:
    pvalue = stats.f_oneway(dataset_mat[dataset_mat[column]==0]["G3"],
                           dataset_mat[dataset_mat[column]==1]["G3"])[1] #Segunda posición de la lista el pvalue
    
    df_binary = df_binary.append(pd.DataFrame({'variable':[column],
                                              'pvalue':[pvalue]}))
    
df_binary = df_binary[df_binary['pvalue']<=threshold]
list_binary_chosen = list(df_binary['variable'].values)
    
df_binary
list_binary_chosen
```

```python
#Agregar el trabajo de las madres
list_multi_chosen=['Mjob']
```

```python
list_all_chosen = list_ordinal_chosen + list_binary_chosen + list_multi_chosen
list_all_chosen
```

# 5. Construcción del modelo

```python
df_mat = dataset_mat[list_all_chosen + ['G3']] # G3 es la variable de salida.. A predecir
df_mat.head()
```

```python
# onehot encoding, para representar los trabajados de las madres. La categoria 5 se representa con 4 ceros
# Recuerden que son 5 categorias. con la combinación de un 1 en la posición, se le da al valor.
aux = pd.get_dummies(df_mat['Mjob'], drop_first = True) #
aux.columns = ['Mjob_' + str(col) for col in list(aux.columns)] #Concatene un sufijo al nombre de la columna de aux
df_mat = df_mat.join(aux, how="inner").drop(columns=['Mjob']) # Se descarta la columna original
df_mat.head()
```

```python
# Normalizado o Escalado de las variables numerales ordinales, las pone en valores entre 0 y 1
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_mat[list_ordinal_chosen] = scaler.fit_transform(df_mat[list_ordinal_chosen]) # devuelve los valores transformados
df_mat.head()
```

```python
# Conversion a variable categorica para que se use un algoritmo de clasificación
mask = df_mat["G3"] <= 10
column_name = 'G3'
df_mat.loc[mask, column_name] = 0

mask = df_mat["G3"] > 10
column_name = 'G3'
df_mat.loc[mask, column_name] = 1

df_mat.head()
```

```python
# Separacion en variables entrada/salida
X = df_mat.drop(columns=["G3"]).values
y = df_mat["G3"].values

# Train/Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
#SVM (Kernel = rbf)
from sklearn.svm import SVC
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score

# Filtered
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print(cm)

print("precision:", precision, " recall:", recall, " f1:", f1)
```
